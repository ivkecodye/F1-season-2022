```python
# Libraries and Dataset

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
%matplotlib inline
dataset=pd.read_csv('Saudi Arabian GP.csv')
dataset.head(20)
```

```python
# X = Drivers ID, y = Final Position, Z = Constructors ID, H = Grid

X = dataset.iloc[:-6,2:-6].values
y = dataset.iloc[:-6,:-8].values
Z = dataset.iloc[:-6,4:-4].values
H = dataset.iloc[:-6,8:].values
```

```python
# Machine Learning Regression Grid vs Final Position

from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(H, y)
```

```python
# Saudi Arabian GP Driver Position

plt.figure(figsize=(15,10))
plt.grid(color='green', linestyle='--', linewidth=0.5)
plt.scatter(X, y, color='blue')
plt.title('Saudi Arabian GP Driver Positions')
plt.xlabel('Driver ID')
plt.ylabel('Position')
```

```python
# Saudi Arabian GP Constructor Position

plt.figure(figsize=(15,10))
plt.grid(color='green', linestyle='--', linewidth=0.5)
plt.scatter(Z, y, color='blue')
plt.title('Saudi Arabian GP Constructor Positions')
plt.xlabel('Constructor ID')
plt.ylabel('Position')
```

```python
# Saudi Arabian GP Driver points

f, ax = plt.subplots(figsize=(15,10))
sns.barplot(x='points', y='driver', data=dataset)
plt.title('Saudi Arabian GP Driver Points')
plt.xlabel('GP Points')
plt.ylabel('Drivers')
```

```python
# Saudi Arabian GP Constructor Points

f, ax = plt.subplots(figsize=(15,10))
sns.barplot(x='points', y='constructor', data=dataset)
plt.title('Saudi Arabian GP Constructor Points')
plt.xlabel('GP Points')
plt.ylabel('Constructors')
```

```python
# Saudi Arabian GP Grid VS Final Position

plt.figure(figsize=(15,10))
plt.grid(color='green', linestyle='--', linewidth=0.5)
plt.scatter(H, y, color='red')
plt.plot(H, regressor.predict(H), color='blue')
plt.title('Grid VS Finish')
plt.xlabel('Start Position')
plt.ylabel('Finish')
```
